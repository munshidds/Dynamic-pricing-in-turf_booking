```python

import pandas as pd
import random

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
times = ["Morning", "Afternoon", "Evening"]
customer_levels = ["Regular", "Premium", "Rare"]
temperatures = range(20, 40)
turf_ratings = range(1, 6)
discount_rates = [0.0, 0.05, 0.08, 0.1, 0.12, 0.15, 0.2]
weather_condition =['Partly Cloudy', 'Mostly Cloudy', 'T-Storms', 'Scattered T-Storms']


# Generate 4000 records
records = []
for _ in range(4000):
    record = {
        "day": random.choice(days),
        "time": random.choice(times),
        "customer_level": random.choice(customer_levels),
        "temperature": random.choice(temperatures),
        "turf_rating": random.choice(turf_ratings),
        "discount_rate": random.choice(discount_rates),
        "weather_condition":random.choice(weather_condition)
    }
    records.append(record)

df = pd.DataFrame(records).drop(['discount_rate'],axis=1)

df.to_csv("turf_booking_data.csv", index=False)
```

```python
data=pd.read_csv('turf_booking_data.csv')
data
```

```python
replace_dic={'day' :{"Monday":2, "Tuesday":2, "Wednesday":2, "Thursday":2, "Friday":1, "Saturday":0, "Sunday":0},
'time':{"Morning":1, "Afternoon":2, "Evening":0},
'customer_level':{"Regular":1, "Premium":2, "Rare":0},
'turf_rating' :{1:5,2:4,4:2,5:1},
'weather_condition':{'Partly Cloudy':0, 'Mostly Cloudy':0, 'T-Storms':1, 'Scattered T-Storms':1}}


new_data=data.replace(replace_dic)

new_data["temperature"]=data["temperature"].apply(lambda x: 0 if x > 35  else 1)
new_data['turf_rating']=new_data['turf_rating']/new_data['turf_rating'].max()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data['discount']=new_data['discount']
data
```

```python
replace_dic={'day' :{"Monday":2, "Tuesday":2, "Wednesday":2, "Thursday":2, "Friday":1, "Saturday":0, "Sunday":0},
'time':{"Morning":1, "Afternoon":2, "Evening":0},
'customer_level':{"Regular":1, "Premium":2, "Rare":0},
'weather_condition':{'Partly Cloudy':0, 'Mostly Cloudy':0, 'T-Storms':1, 'Scattered T-Storms':1}}


data=data.replace(replace_dic)
```

```python
X=data.drop(['final price',"discount"],axis=1)
y=data.discount
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# model=LinearRegression()
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()
rf_model.fit(X_train,y_train)
y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
